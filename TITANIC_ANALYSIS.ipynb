```python
#importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#assign a variable to dataset
df = pd.read_csv('/content/train.csv')
```

```python
#understanding the dataset
df.head(5)
```

```python
#drop cabin column from datset
df.drop('Cabin', axis=1,inplace=True)

#display output
df.head()
```

```python
#drop name and ticket column
df.drop(columns=['Name','Ticket'],inplace=True)

#display output
df.head()
```

```python
#find the mode of  Embarked column
m=df['Embarked'].mode()
m
#fill missing embarked values
df['Embarked'].fillna(value='S',inplace=True)

#display output
df.head(10)
```

```python
df.isna().sum()
```

```python
#find median age in the Pclass
df.groupby('Pclass')['Age'].median()
```

```python
#use the median age in Pclass to fill missing values
df['Age'].fillna(df.groupby('Pclass')['Age'].transform('median'),inplace=True)

#display output
df.head()
```

```python
df.isnull().sum()
```

```python
#encode the sex column to binary
df['Sex'] = df['Sex'].map({'male':0,'female':1})

#display output
df.head()
```

```python
#encode the Embarked column to numerical
df['Embarked'] = df['Embarked'].map({'S':0,'C':1,'Q':2})

#display output
df.head(40)
```

```python
# create pie chart using the Survived and Age
df.groupby('Survived')['Age'].mean().plot.pie(autopct='%1.1f%%')
```

```python
#cerate piechart using the survived and sex
df.groupby('Survived')['Sex'].value_counts().plot.pie(autopct='%1.1f%%')
```

```python
#drop name and ticket column
del df ['PassengerId']
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Assuming 'Survived' is the target variable and the rest are features
X = df.drop('Survived', axis=1)  # Features
Y = df['Survived']  # Target variable

# Now you can split the data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)


#print x train
X_train
```

```python
#print x test
X_test
```

```python
#print Y-train
Y_train
```

```python
#feature scaling of x train
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)


#print output
X_train
```

```python
# Using logistics regression for machine learning
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
log.fit(X_train,Y_train)
```

```python
log.predict(X_test)
```

```python
# checking the accuracy of prediction
log.score(X_test,Y_test)*100
```
